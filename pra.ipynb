### 지갑 주소 입력받아서 리플 잔액 조회
- Ledger 직접 조회

```python
import requests

def get_xrp_balance(address: str):
    """
    리플 지갑 주소(XRP Ledger address)를 입력받아,
    해당 지갑이 보유한 XRP 잔액을 반환하는 함수
    """
    url = "https://s1.ripple.com:51234/"
    headers = {"Content-Type": "application/json"}

    data = {
        "method": "account_info",
        "params": [
            {
                "account": address,
                "ledger_index": "validated",
                "strict": True
            }
        ]
    }

    try:
        response = requests.post(url, json=data, headers=headers)
        response.raise_for_status()
        result = response.json()

        # 정상 응답인지 확인
        if "result" in result and "account_data" in result["result"]:
            balance_drops = int(result["result"]["account_data"]["Balance"])
            balance_xrp = balance_drops / 1_000_000  # drops → XRP 단위 변환
            return balance_xrp
        else:
            return "지갑 주소를 찾을 수 없습니다."

    except requests.exceptions.RequestException as e:
        return f"요청 에러: {e}"
    except Exception as e:
        return f"예외 발생: {e}"
```

```python
from web3 import Web3
import requests

# ① 리플 원장에서 XRP 조회
def get_xrp_balance(address: str):
    url = "https://s1.ripple.com:51234/"
    headers = {"Content-Type": "application/json"}
    data = {
        "method": "account_info",
        "params": [{"account": address, "ledger_index": "validated", "strict": True}]
    }
    try:
        r = requests.post(url, json=data, headers=headers).json()
        balance_drops = int(r["result"]["account_data"]["Balance"])
        return balance_drops / 1_000_000
    except Exception:
        return 0.0

# ② EVM 네트워크에서 mXRP 조회
def get_mxrp_balance(wallet_address: str):
    # BNB Chain 메인넷 예시 RPC
    rpc_url = "https://bsc-dataseed.binance.org/"
    web3 = Web3(Web3.HTTPProvider(rpc_url))
    
    # mXRP 토큰 컨트랙트 주소 (예시 — 실제 주소 확인 필요)
    contract_address = Web3.to_checksum_address("0x6fA...")  # 실제 mXRP 주소로 변경
    abi = [
        {"constant": True, "inputs": [{"name": "_owner", "type": "address"}],
         "name": "balanceOf", "outputs": [{"name": "balance", "type": "uint256"}],
         "type": "function"}
    ]
    contract = web3.eth.contract(address=contract_address, abi=abi)
    balance = contract.functions.balanceOf(wallet_address).call()
    return balance / (10 ** 18)  # 일반 ERC-20은 소수점 18자리

# ③ 통합 조회 함수
def check_balances(xrp_addr, evm_addr):
    xrp = get_xrp_balance(xrp_addr)
    mxrp = get_mxrp_balance(evm_addr)
    return {"XRP": xrp, "mXRP": mxrp}
```

```python
import requests
import json
from datetime import datetime

def get_xrpl_ledger_info(ledger_index="validated"):
    """
    리플(XRPL) 원장 구조를 조회하는 함수.
    기본은 최신 검증 완료된 원장(validated)을 가져옴.
    """
    url = "https://s1.ripple.com:51234/"
    headers = {"Content-Type": "application/json"}
    data = {
        "method": "ledger",
        "params": [{
            "ledger_index": ledger_index,
            "accounts": False,
            "full": False,
            "transactions": True,
            "expand": True
        }]
    }

    try:
        r = requests.post(url, json=data, headers=headers, timeout=10)
        r.raise_for_status()
        result = r.json()["result"]

        ledger = result["ledger"]
        info = {
            "ledger_index": ledger["ledger_index"],
            "ledger_hash": ledger["ledger_hash"],
            "close_time": datetime.utcfromtimestamp(ledger["close_time"] + 946684800).strftime("%Y-%m-%d %H:%M:%S UTC"),
            "total_coins": int(ledger["total_coins"]) / 1_000_000,
            "transaction_count": len(ledger.get("transactions", [])),
            "parent_hash": ledger["parent_hash"],
        }
        return info

    except Exception as e:
        return {"error": str(e)}
```

```python
info = get_xrpl_ledger_info()
print(json.dumps(info, indent=4))
```

```python
import requests
import json

def get_recent_xrp_transactions(limit=50):
    """
    XRPL에서 최근 거래(limit 개) 조회
    """
    url = "https://s1.ripple.com:51234/"
    headers = {"Content-Type": "application/json"}

    # 먼저 최신 원장 번호 가져오기
    ledger_request = {
        "method": "ledger",
        "params": [{"ledger_index": "validated", "transactions": True, "expand": True}]
    }

    try:
        r = requests.post(url, json=ledger_request, headers=headers, timeout=10)
        r.raise_for_status()
        ledger = r.json()["result"]["ledger"]

        txs = ledger.get("transactions", [])

        # 최근 거래 limit만큼 자르기
        recent_txs = txs[-limit:] if len(txs) > limit else txs
        return recent_txs

    except Exception as e:
        return {"error": str(e)}

# 사용 예시
if __name__ == "__main__":
    transactions = get_recent_xrp_transactions(50)
    print(json.dumps(transactions, indent=2))
```
